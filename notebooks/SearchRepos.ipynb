```python
search_dir = Path("../telegram-ml-contest-data/")

langs = [
    "Bison",
    "Protobuf",
    "Smalltalk",
    "Ada",
    "ActionScript",
    "ASP",
    "Delphi",
    "Pascal",
    "AWK",
    "D",
    "Forth",
    "Scheme",
    "VDScript",
    "IDL",
    "Basic",
    "Tcl",
    "Wolfram",
    "QML",
    "GraphQL",
    "AutoHotKey",
    "Raku",
    "AppleScript",
    "Vala",
    "Elm",
    "NGINX",
    "Textile",
    "Gradle",
    "UnrealScript",
    "Regex",
    "Hack",
    "TL",
    "Crystal",
    "SAS",
    "Nim",
    "GAMS",
    "ABAP"
]

# langs = {
#     "1C Enterprise",
#     "FIFT",
#     "OpenEdgeABL",
#     "Lisp",
#     "PLSQL",
#     "F#",
#     "Solidity",
#     "Icon",
#     "FunC",
#     "C#",
#     "CommonLisp",
#     "C++",
#     "Keyman",
#     "Logo",
#     "APEX",
# }

USERNAME = 'kst179'
TOKEN = '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'

def authenticate_github():
    session = requests.Session()
    session.auth = (USERNAME, TOKEN)
    return session

session = authenticate_github()

url = "https://api.github.com/search/repositories?q=language:\"{lang}\"&per_page=100&page={page}"

fails = {}
lang_repos = defaultdict(list)

for lang in tqdm(langs):
    # lang_dir = search_dir / lang 
    # lang_dir.mkdir(exist_ok=True, parents=True)

    for page in tqdm(range(1, 11), leave=False):
        timeout = 61
        retries = 10
        while True:
            response = session.get(url.format(lang=lang, page=page))
            
            if response.status_code == 403:
                time.sleep(timeout)
                timeout = 1
                retries -= 1
            else:
                break

            if retries == 0:
                break
            
        if not response.ok:
            fails[(lang, page)] = response.text
            continue

        repos_info = response.json()
        for repo in repos_info["items"]:
            lang_repos[lang].append(repo["full_name"])
        
    with open(search_dir / "repos2.json", "w", encoding="utf8") as file:
        json.dump(lang_repos, file, indent=4, ensure_ascii=False)
```
