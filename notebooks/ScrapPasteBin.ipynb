```python
import requests
import bs4

response = requests.get("https://pastebin.com/archive")
```

```python
bs = bs4.BeautifulSoup(response.text)
```

```python
table = bs.find("table", attrs={"class": "maintable"}).findChild("tbody")

for item in table.children:
    url = item.find("a")
    break
```

```python
table.find_all("a")
```

```python
url.attrs["href"]
```

```python
pastes = []
langs = []
names = []

findall = table.find_all("a")

for url, lang in zip(findall[::2], findall[1::2]):
    name = url.text
    url = url.attrs["href"]
    lang = lang.text
    
    response = requests.get("https://pastebin.com/raw{url}")
    pastes.append(response.text)
    langs.append(lang)
    names.append(name)
```
