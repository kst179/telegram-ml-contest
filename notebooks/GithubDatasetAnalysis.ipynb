```python
from pathlib import Path
import json

data = Path("github_data")

for subdir in data.iterdir():
    if subdir.is_dir():
        n_items = len(list(subdir.iterdir()))
        print(subdir, n_items)
        
    else:
        with open(subdir, encoding="utf8") as file:
            code = json.load(file)
            
        print(subdir, len(code))
```
